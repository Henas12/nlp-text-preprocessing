## Tokenization

is a process of coverting texts(such as paragraph, sentence) into smaller units called Token(words, characters, subwords)

```python
import nltk
nltk.download('punkt_tab')
```

```python
corpus = """"NLTK, or the Natural Language Toolkit, is a Python library or Natural Language Processing (NLP).
It provides a wide range tools and resources.
"""
```

```python
#Tokenization
# Paragraphs --> Sentences

from nltk import sent_tokenize
sentences = nltk.sent_tokenize(corpus)
for sentence in sentences:
    print(sentence)
```

```python
## Tokenization
## Paragraph --> Words
## Sentence --> words

from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentence in sentences:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```
