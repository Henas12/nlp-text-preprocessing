# Stemming:
is a proccess of reducing a word to its stem



```python
words = [
    "running", "runner", "runs", "easily", "fairness",
    "studying", "studies", "studied", "happiness", "unhappily",
    "cats", "boxes", "playing", "played", "plays",
    "movement", "agreement", "flying", "flies", "denying",
    "denial", "better", "worse", "eating", "eaten"
]
```

###PorterStemmer

```python
#ProterStemmer

from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()

for word in words:
  print(f'{word} ---> {stemmer.stem(word)}')
```

##**RegExpStemmer**

```python
#RegexpStemmer allows to filter out some unwated letters.
from nltk.stem import  RegexpStemmer

regexp_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
regexp_stemmer.stem('running'), regexp_stemmer.stem('ingrunning')
```

##SnowBall Stemmer

```python
from nltk.stem import SnowballStemmer
```

```python
snowball_stemmer = SnowballStemmer(language='english')
```

```python
for word in words:
  print(f'{word} ---> {snowball_stemmer.stem(word)}')
```

```python
stemmer.stem('fiarly'), snowball_stemmer.stem('fiarly')
```


# Word Lemmatization:
is a technique like stemming, but lemmetization will give a valid word that means the same meaning


```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer  = WordNetLemmatizer()
```

```python
for word in words:
  print(f'{word} ---> {lemmatizer.lemmatize(word)}')
```

```python
for word in words:
  print(f'{word} ----> {lemmatizer.lemmatize(word, pos="a")}')

```
